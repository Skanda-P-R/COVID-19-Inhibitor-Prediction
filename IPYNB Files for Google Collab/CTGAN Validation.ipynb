```python
import pubchempy as pcp
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['CTGAN Syntetic Data2.csv']))
```

```python
def is_valid_cid(cid):
    try:
        compound = pcp.get_compounds(cid, 'cid')
        return True
    except pcp.PubChemHTTPError:
        return False


df['ValidCID'] = df['CID'].apply(is_valid_cid)
```
