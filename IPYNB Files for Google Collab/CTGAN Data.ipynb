```python
import pubchempy as pcp
import pandas as pd

from google.colab import files
uploaded = files.upload()
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['DDH Data.csv']))
```

```python
df.head()
```

```python
data = []

for i in df['SMILES']:
    props = pcp.get_properties(['MolecularFormula'], i, 'smiles')
    data.append(props)
```

```python
rows = []
columns = data[0][0].keys()
for i in range(104):
    rows.append(data[i][0].values())
props_df = pd.DataFrame(data=rows, columns=columns)
props_df.head()
```

```python
props_df['pIC50'] = df['pIC50 (IC50 in microM)']
props_df.drop(columns=['MolecularFormula'])
props_df.head()
```

```python
props_df.drop(columns=['MolecularFormula'])
```

```python
props_df.to_csv('DDH Data with Properties.csv')
```

```python
from google.colab import files
props_df.to_csv('DDH Data with Properties.csv')
files.download('DDH Data with Properties.csv')
```
